```python
import pandas as pd
import requests

# Make the API request
response = requests.get("https://api.currencyfreaks.com/v2.0/rates/latest?apikey=54cae95200774f9a97752166ddfcd4f6")

# Convert the 'rates' dictionary to a DataFrame
df = pd.DataFrame(response.json()['rates'].items(), columns=['Currency', 'Exchange Rate(to_INR)'])

# Show the DataFrame
print(df)
```

```python
df.to_csv('currency_exchange_rates.csv', index=False)
```
